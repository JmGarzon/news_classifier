```python
import pandas as pd
import numpy as np

# leer el archivo noticias.csv
noticias = pd.read_csv('/content/drive/MyDrive/7. EAFIT/Algebra/data/noticias.csv')
noticias.shape
```

#2 Preparación del texto:
Tokenización, eliminación de stopwords, y lematización

```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

def clean_text(text):
  # convertir a minúsculas
  text = text.lower()
  # reemplazar caracteres especiales y números por espacios
  text = re.sub('[^a-zA-ZñÑ]', ' ', text)
  # tokenizar
  words = word_tokenize(text)
  # eliminar stopwords y lematizar
  lemmatizer = WordNetLemmatizer()
  words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('spanish') and len(word) >  1]
  # unir palabras en un solo texto
  text = ' '.join(words)
  return text

# limpiar noticias.csv
noticias['clean_text'] = noticias['news_text_content'].apply(clean_text)
```
